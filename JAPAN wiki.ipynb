```python
from bs4 import BeautifulSoup
import requests

url='https://en.wikipedia.org/wiki/List_of_largest_Japanese_companies'

r = requests.get(url)
soup = BeautifulSoup(r.text, 'html.parser')
```

```python
print(soup)
```

```python
soup.find('table')
```

```python
soup.find_all('table')[0]
```

```python
soup.find('table', class_='wikitable sortable')
```

```python
table = soup.find_all('table')[0]
print(table)
```

```python
Japan_titles = table.find_all('th')
Japan_titles
```

```python
Japan_table_titles = [title.text.strip() for title in Japan_titles]

print(Japan_table_titles)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = Japan_table_titles)

df
```

```python
column_data = table.find_all('tr')
```

```python
for row in column_data[1:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]
    print(individual_row_data)
    
    length = len(df)
    df.loc[length] = individual_row_data
```

```python
df
```

```python
df.to_csv("wikiJPN_data.csv",index=False)
```
